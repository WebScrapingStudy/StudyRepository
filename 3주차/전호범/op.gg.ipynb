### 영어로 나오길래 한글로 받을 수 있을까 고민해 보았습니당

```python
from bs4 import BeautifulSoup
import requests
url = "https://www.op.gg/champion/statistics"
headers = {'Content-Language': 'ko-KR','cookie':'customLocale=ko_KR'}
result = requests.get(url = url,headers=headers)
```

* 리퀘스트s 모듈을 사용해서 get 메소드로 패킷을 보냅니닷!  
* 헤더를 수정할수 있어요!! 헤더를 딕셔너리 타입으로 파라미터로 넘겨줍니다.

```python
print(result)
```

```python
soup = BeautifulSoup(result.content,"html.parser")
```

위에거 복붙.,

```python
champ = soup.find("div",{"class":"champion-index__champion-list"}).findAll("a")
link_list = []
for i in champ:
    name = i.find("div",{"class":"champion-index__champion-item__name"})
    link_list.append([name.text,"https://op.gg"+i.get('href')])
```

## 쿠키에 커스텀 로컬 ko_KR 추가후에 한글로 출력 완료

* 헤더는 크롬 개발자 도구에서 네트워크 탭에서 볼수 있습니다.   
![op헤더](op헤더.png)

```python
for i in link_list:
    print(i)
```

### select 메소드도 굉장히 효과적입니다.

```python
tear = soup.select("div.champion-stats-header-info__tier > b")  
print(tear[0].text)
```

```python
from time import sleep
headers = {'Content-Language': 'ko-KR','cookie':'customLocale=ko_KR'}
```

```python
for i in link_list:
    sleep(1)  # 5초 휴식
    result = requests.get(url = i[1],headers=headers) #가렌부터 끝까지 호출
    soup = BeautifulSoup(result.content,"html.parser")  #파싱
    tear = soup.select("div.champion-stats-header-info__tier > b") # 티어 찾기
    i.append(tear[0].text) # 저장
    print(i) #출력
```

### 티어별로 정렬할까..

```python
from operator import itemgetter
tier_list = sorted(link_list,key=itemgetter(2))
```

```python
for i in tier_list:
    print(i)
```
